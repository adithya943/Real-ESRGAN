# Testing Degradations

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

a = cv.imread('/home/adithya/adithya/Real-ESRGAN/data/klothed-oldnavy-images/webcontent_0010_698_786_cn10698786.jpg')
a = cv.cvtColor(a, cv.COLOR_BGR2RGB)
def repeat_resize(img, target_size_factor = 0.5, it = 10):
    orig_size = img.shape
    for i in tqdm(range(it)):
        img = cv.resize(img, (int(orig_size[1]*target_size_factor),int(orig_size[0]*target_size_factor)), cv.INTER_CUBIC)
        img=cv.resize(img,(orig_size[1],orig_size[0]))
    return img
```

```python
degraded = repeat_resize(a, 0.5, 10000)
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,15))
axs[0].imshow(a)
axs[1].imshow(degraded)
```
